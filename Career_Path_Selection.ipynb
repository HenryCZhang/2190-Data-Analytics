Income Rates, Career Satisfaction, and Career Choice Factors Among University Students

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

Career_Path = pd.read_csv('Career_Path_Selection_Student_Responces.csv')
```

```python
career_path_info = Career_Path.info()
print(career_path_info)
```

**1)Â Are you confident in your chosen career path?**

```python
# Count the responses for confidence in chosen career path
confidence_counts = Career_Path['Are you confident in your chosen career path?'].value_counts()


plt.figure(figsize=(8, 6))
Career_Path['Are you confident in your chosen career path?'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    colors=['cornflowerblue', 'salmon'],
    textprops={'fontsize': 14}  # Setting font size for labels
)

plt.title('Proportion of Students Feeling Confident in Their Chosen Career Path', fontsize=16)
plt.ylabel('')  # Remove y-label for aesthetics
plt.show()
```

**2) What factors affect your career decision-making process?**

```python
# Extract and clean the column for factors influencing career decisions
factors_column = Career_Path['What factors affect your career decision-making process?'].dropna()

# Split multiple selections and count occurrences of each factor
factor_counts = factors_column.str.get_dummies(sep=', ').sum()

plt.figure(figsize=(10, 8))
factor_counts.sort_values().plot(kind='barh', color='cornflowerblue')
plt.title("Factors Affecting Students' Career Decision-Making Process", fontsize=16)
plt.xlabel("Count of Students", fontsize=14)
plt.ylabel("Factors", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

The number one factor affecting students' career decision-making process is students' passion for a specific field. This is followed by financial stability and job market trends.

**3)Â How did you choose your career path?**

```python
#Remove the extra space from the column name
Career_Path.columns = Career_Path.columns.str.strip()

# Extract and clean the column for factors influencing career path choices
factors_column = Career_Path['How_did you_choose_your_career_path?'].dropna()

# Split multiple selections and count occurrences of each factor
factor_counts = factors_column.str.get_dummies(sep=', ').sum()

plt.figure(figsize=(10, 8))
factor_counts.sort_values().plot(kind='barh', color='cornflowerblue')
plt.title("Impact of Factors on Students' Career Path Choices", fontsize=16)
plt.xlabel("Count of Students", fontsize=14)
plt.ylabel("Factors Influencing Career Path Choices", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

**4)Â Have you sought advice from a career counselor or career advisor?**

Most students choose their career paths based on personal interests and passion. second major factor is the influence of their family and friends. A few students choose their career paths through professional counseling. (we can talk about why this is the case.)
